### Загружаю библиотеки и данные

```python
print ('Импортируем необходимые библиотеки')

from tensorflow.keras.datasets import mnist #Библиотека с базой Mnist
from tensorflow.keras.models import Sequential #НС прямого распространения
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization #Полносвязные слой
from tensorflow.keras import utils #Утилиты для to_categorical
from tensorflow.keras.preprocessing import image #Для отрисовки изображения
import numpy as np #Библиотека работы с массивами
import matplotlib.pyplot as plt #Отрисовка изображений
from PIL import Image #Отрисовка изображений
from sklearn.model_selection import train_test_split
#from scikit-learn.model_selection import train_test_split

# Импорт библиотек для моего кода
import datetime
import os
import sys
#Отрисовывать изображения в ноутбуке, а не в консоль или файл
%matplotlib inline

#Загрузка данных Mnist

print ('Загружаем данные Mnist в память')
(x_train_org, y_train_org), (x_test_org, y_test_org) = mnist.load_data()
print ('Ok')
```

### Подготавливаю data set для анализа и проверяю данные

```python
x_train = x_train_org.reshape(60000, 784)
x_test = x_test_org.reshape(10000, 784)
y_train = utils.to_categorical(y_train_org, 10)
y_test = utils.to_categorical(y_test_org, 10)
# Нормализация данных
x_train = x_train.astype('float32')
x_train = x_train / 255
x_test = x_test.astype('float32')
x_test = x_test / 255
```

```python
n_pic = 221 #Номер картинки

print(y_test_org[n_pic])
print(y_test[n_pic])

print ('Показываем произвольную картинку номер ', n_pic)
print (y_train [n_pic])
plt.imshow(Image.fromarray(x_train_org[n_pic]).convert('RGBA'))
plt.show()
```

```python
print ('Shapes of massives')
print ('x_train = ', x_train.shape)
print ('y_train = ', y_train.shape)
print ('x_test = ', x_test.shape)
print ('y_test = ', y_test.shape)
print ('x_train_org = ', x_train_org.shape)
print ('y_train_org = ', y_train_org.shape)
```

### Моя программа генерации и обучения нейронной сети

```python
# Процедура создания и обучения сети по параметрам:

# nc1 - Число нейронов первого слоя
# nc2 - Число нейронов второго слоя
# nc3 - Число нейронов третьего слоя
# fact_in - активационная функция скрытых слоёв
# fact_out - активационная функция выходного слоя

# Глобальные переметры сети:
# n_epo - число эпох
# batch - batch_size
# split - validation_split

def createNet (my_x_train, my_y_train, nc1, nc2, nc3, fact_in, fact_out):
    model = Sequential()
    model.add(Dense(nc1, input_dim=784, activation=fact_in))
    if nc2 > 0: model.add(Dense(nc2, activation=fact_in))
    if nc3 > 0: model.add(Dense(nc3, activation=fact_in))
    model.add(Dense(10, activation=fact_out))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    
    # Запускаем сеть на обучение

    mod_history = model.fit(my_x_train, my_y_train, batch_size=batch, epochs=n_epo, verbose=0, validation_split=split)
    return model, mod_history
```

### Набор параметров для тестирование сети

```python
test_exa = [
    [400, 0, 0, "relu", "softmax"],
    [800, 400, 0, "relu", "softmax"],
    [800, 800, 400, "relu", "softmax"]
    ]
n_epo = 20; batch = 128; split = 0.2;

# Массив x для показа графиков по эпохам
x = []
for i in range (0, n_epo):
  x.append (i+1)
```

### Обучаем на первых 500 примерах

```python
n_examples = 500
print ('Тестируем ', n_examples, ' примеров')
x_train_new = x_train [0:n_examples]
y_train_new = y_train [0:n_examples]
print (x_train_new.shape) 
print (y_train_new.shape)

#nom - номер сети в списке acc_sum, которую тестируем
nom = 0;

for testa in test_exa:
    print ("Создаем сеть номер ", nom+1, ": с параметрами: ", test_exa [nom], " => ", end='')
    nom += 1
    # Время перед началом создания и обучения сети
    dat1 = datetime.datetime.now()

    my_model, mh = createNet (x_train_new, y_train_new, testa[0], testa[1], testa[2], testa[3], testa[4])

    # Время обучения сети
    delta = datetime.datetime.now() - dat1 

    print (mh.history ["accuracy"][-1], "  время = ", delta.total_seconds(), ' сек')

    scores = my_model.evaluate(x_test, y_test, verbose=0)
    print("Доля верных ответов на тестовых данных:     ", round(mh.history ['val_accuracy'][-1] * 100, 4), "%", sep="")
    print("Доля верных ответов на проверочной выборке: ", round(scores[1] * 100, 4), "%", sep="")
   
print ('Finished')
plt.plot (x, mh.history ["accuracy"], mh.history ["val_accuracy"])
plt.show()
plt.plot (x, mh.history ["loss"], mh.history ["val_loss"])
```
